```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Impact_of_Mobile_Phone_on_Students_Health.csv")
```

```python
# AGE DISTRIBUTION
plt.figure(figsize=(10,8))
sns.histplot(data=df,x='Age',bins=10,kde=True)
plt.title('Age Distribution Of Students')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# STUDENTS PERFORMANCE IMPACTS DUE TO MOBILE PHONE USE
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='Performance impact', order=['Strongly agree','Agree','Neutral'])
plt.title('STUDENTS PERFORMANCE IMPACTS DUE TO MOBILE PHONE USE') 
plt.xlabel('Performance impact')
plt.ylabel('Count')
plt.show()              
```

```python
#Health Ratings
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='Health rating', order=['Excellent', 'Good', 'Fair', 'Good;Fair', 'Excellent;Good', 'Poor', 'Excellent','Good','Fair', 'Poor'])
plt.title('Health Distribution') 
plt.xlabel('Health rating')
plt.ylabel('Count')
plt.show() 
```
